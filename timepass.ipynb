```python
import pandas as pd
import re
```

```python
text = "Hello Chat! How's it going?"
```

```python
df = pd.Series(text.split())
```

```python
print(df.value_counts)
```

```python
def count(text):
    text = re.findall(r'\b\w+\b', text.lower())
    df = pd.Series(text)
    print(df.value_counts())
```

```python
text = "Hello Chat! How's it going?"
```

```python
count(text)
```
